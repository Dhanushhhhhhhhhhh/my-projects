```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 5))
sns.boxplot(y=df['Exchange'], color='skyblue')
plt.title("Outlier Detection in Exchange Rates")
plt.ylabel("Exchange Rate")
plt.grid(True)
plt.show()
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
outliers = detect_outliers_iqr(df, 'Exchange')
print(outliers)
```

```python

def optimize_dataframe(df):
    """
    Convert int64 to int32, float64 to float32, and object to category using a for loop.
    
    Parameters:
    df (pd.DataFrame): Input DataFrame to optimize.

    Returns:
    pd.DataFrame: Optimized DataFrame with reduced memory usage.
    """
    for col in df.columns:
        if df[col].dtype == "int64":
            df[col] = df[col].astype("int32")  # Convert int64 → int32
        elif df[col].dtype == "float64":
            df[col] = df[col].astype("float32")  # Convert float64 → float32
        elif df[col].dtype == "object":
            df[col] = df[col].astype("category")  # Convert object → category
    return df
```
